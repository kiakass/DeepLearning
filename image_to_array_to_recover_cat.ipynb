```
img_path = '/content/gdrive/MyDrive/04.Colab Python/DATA/MNIST/0/'
img_name = 'cat_2.jpg'
img_nm=img_path+img_name
```

```
import matplotlib.pyplot as plt
import matplotlib.image as img

img_nm2 = img.imread(img_path+img_name)
plt.imshow(img_nm2)
```

```
# Preprocess the image into a 4D tensor using keras.preprocessing

from keras.preprocessing import image

img = image.load_img(img_nm, target_size=(500,500))

img_tensor = image.img_to_array(img)
img_tensor.shape
```

```
 # expand a dimension (3D -> 4D)

img_tensor = np.expand_dims(img_tensor, axis=0)

img_tensor.shape

 # scaling into [0, 1]

img_tensor /= 255.
```
