```python
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib import style
from sympy import Derivative, symbols
from scipy.misc import derivative
from sympy import *
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_dev(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
def tanh(x):
    return np.tanh(x)

def tanh_dev(x):
    return 1.0 - np.tanh(x)**2
```

```python
def ReLU(x):
    return x * (x > 0)

def ReLU_dev(x):   
    return 1*(x>0)
```

```python
def LeakyReLU(x):
    a=0.1
    return list(map(lambda x: x if x > 0 else  a*x,x))

def LeakyReLU_dev(x):
    a=0.1
    return list(map(lambda x: 1 if x > 0 else  a,x))
```

```python
# ELU
def ELU(x):
    a=0.1
    return list(map(lambda x: x if x > 0 else  a*(np.exp(x)-1),x))

def ELU_dev(x): 
    a=0.1
    return list(map(lambda x: 1 if x > 0 else  a*(np.exp(x)-1)+a,x))
```

```python
# Graph 그리기
def g_active(name,name2,xlim,ylim):
    
    style.use('ggplot')
    x=np.linspace(-xlim, xlim, 100)
    
    plt.figure(figsize=(10,5))

    plt.subplot(121)
    pdf = name(x)
    title=str(name).split(' ')
    plt.title(title[1]) 
    plt.plot(x,pdf)
    
    plt.subplot(122)
    pdf_d = name2(x)
    plt.ylim(0,ylim)
    plt.title(title[1] +' Derivative') 
    f,=plt.plot(x,pdf_d,color='steelblue')
```

```python
#함수실행 : sigmoid, tanh, ReLU, ELU, 
g_active(sigmoid,sigmoid_dev,4,2) 
g_active(tanh,tanh_dev,4,2) 
g_active(ReLU,ReLU_dev,4,2) 
g_active(LeakyReLU,LeakyReLU_dev,4,2)  #fuction, derivate, xlim, ylim
g_active(ELU,ELU_dev,4,2) 
```
